```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
# fig is the figure; ax is the axes--useful for annotation
fig, ax = plt.subplots()
# this plots data from x and y axes
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()--alt method but i hate it
# creates empty graph
fig = plt.figure()
ax = fig.add_subplot()
# adds data to graph
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
# show() displays active figure object and opens window to display if i have 2+ sets of data to plot
# show() can cause problems if used more than once per session, have to restart kernel
plt.show()
```

```python
# create plot
# transparency is alpha
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], color='skyblue', alpha=.8, edgecolors='black', linewidths=2, label='Boston')

# Create labels for the x and y axes.
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.invert_yaxis()
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add the legend.
ax.legend()
```

```python
plt.pie(y_axis, labels=x_axis)

plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# setting a value to offset the biggest wedge in our data, july
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.subplots(figsize=(8, 8))
# autopct adds percents on wedges
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# startangle rotates pie chart
ax.pie(y_axis,
       labels=x_axis,
       explode=explode_values,
       autopct='%.1f%%',
       shadow=True,
       startangle=265,
       colors=colors)
ax.invert_yaxis()
plt.show()
```
