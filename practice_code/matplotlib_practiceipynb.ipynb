```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
# fig is the figure; ax is the axes--useful for annotation
fig, ax = plt.subplots()
# this plots data from x and y axes
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()--alt method but i hate it
# creates empty graph
fig = plt.figure()
ax = fig.add_subplot()
# adds data to graph
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
# show() displays active figure object and opens window to display if i have 2+ sets of data to plot
# show() can cause problems if used more than once per session, have to restart kernel
plt.show()
```
